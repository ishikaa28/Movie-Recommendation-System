```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector=cv.fit_transform(d1['tags'].values.astype('U')).toarray()
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim= cosine_similarity(vector)
```

```python
sim
```

```python
d1[d1['title']== "The Godfather"].index[0]
```

```python
dist=sorted(list(enumerate(sim[2])), reverse=True, key= lambda vector:vector[1])
for i in dist[0:5]:
  print(d1.iloc[i[0]].title)
```

```python
def recommend(data):
  index=d1[d1['title']==  data].index[0]
  dist=sorted(list(enumerate(sim[index])), reverse=True, key= lambda vector:vector[1])
  for i in dist[0:5]:
    print(d1.iloc[i[0]].title)

```

```python
recommend("Iron Man")
```
